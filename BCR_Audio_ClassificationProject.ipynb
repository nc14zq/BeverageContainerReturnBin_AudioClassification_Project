#**2 Data Preparation**

**2.1 Data Extraction**

```python
import soundfile as sf
from scipy.signal import find_peaks

SR = 16000
RMS_FRAME = 512
RMS_HOP = 128
PRE_MS = 150
POST_MS = 850
MIN_GAP_MS = 1000

CAN_THRESHOLD = 0.14
BOTTLE_THRESHOLD = 0.20

FILES = [
    "Can_Train1.m4a",
    "Can_Train2.m4a",
    "Can_Train3.m4a",

    "Can_Validation.m4a",
    "Can_Validation2.m4a",
    "Can_Test.m4a",
    "Can_Test2.m4a",

    "Bottle_Train1.m4a",
    "Bottle_Train2.m4a",
    "Bottle_Train3.m4a",

    "Bottle_Validation.m4a",
    "Bottle_Validation2.m4a",
    "Bottle_Test.m4a",
    "Bottle_Test2.m4a",

]

def load_and_normalize(audio_path):
    y, _ = librosa.load(audio_path, sr=SR, mono=True)
    peak = np.max(np.abs(y))
    if peak > 0:
        y = y / peak
    return y

def get_rms_envelope(y):
    rms = librosa.feature.rms(
        y=y,
        frame_length=RMS_FRAME,
        hop_length=RMS_HOP
    )[0]
    times = librosa.frames_to_time(
        np.arange(len(rms)),
        sr=SR,
        hop_length=RMS_HOP,
        n_fft=RMS_FRAME
    )
    return rms, times

def detect_impacts(y, threshold):
    rms, times = get_rms_envelope(y)
    min_gap_frames = int((MIN_GAP_MS / 1000.0) * SR / RMS_HOP)

    peaks, _ = find_peaks(
        rms,
        height=threshold,
        distance=min_gap_frames,
        prominence=0.02
    )
    return rms, times, peaks

def parse_file_info(fname):
    label = "bottle" if "Bottle" in fname else "can"

    if "Train" in fname:
        split = "train"
    elif "Validation" in fname:
        split = "validation"
    elif "Test" in fname:
        split = "test"
    else:
        split = "unknown"

    return split, label

def get_threshold(fname):
    if "Bottle" in fname:
        return BOTTLE_THRESHOLD
    elif "Can" in fname:
        return CAN_THRESHOLD
    else:
        return 0.15

def extract_impacts_from_file(fname):
    if not os.path.exists(fname):
        print(f"Missing file: {fname}")
        return

    threshold = get_threshold(fname)
    split, label = parse_file_info(fname)
    y = load_and_normalize(fname)

    out_dir = os.path.join("extracted_audiofinal", split, label)
    os.makedirs(out_dir, exist_ok=True)

    rms, times, peaks = detect_impacts(y, threshold)

    pre_samples = int(PRE_MS / 1000.0 * SR)
    post_samples = int(POST_MS / 1000.0 * SR)
    target_len = pre_samples + post_samples

    base_name = os.path.splitext(os.path.basename(fname))[0]
    saved = 0

    for i, p in enumerate(peaks):
        peak_sample = int(times[p] * SR)
        start = max(0, peak_sample - pre_samples)
        end = min(len(y), peak_sample + post_samples)

        clip = y[start:end]
        if len(clip) < target_len:
            clip = np.pad(clip, (0, target_len - len(clip)))

        out_path = os.path.join(out_dir, f"{base_name}_impact_{i+1:03d}.wav")
        sf.write(out_path, clip, SR)
        saved += 1

    print(f"{fname}: saved {saved} clips -> {out_dir}")

for fname in FILES:
    extract_impacts_from_file(fname)
```

**2.2 Data Cleaning**

```python
import glob
import shutil

SR = 16000

BASE_DIR = "extracted_audiofinal"
REJECT_DIR = "rejected_audiofinal"

#Split Size Limits
SPLIT_LIMITS = {
    "train": 400,
    "validation": 120,
    "test": 120,
}

#Valid audio conditions tuning
PRE_MS = 150
CHECK_START_MS = 40
CHECK_END_MS = 500
MIN_POST_RMS = 0.035
MIN_ACTIVE_RATIO = 0.12
ACTIVE_AMP = 0.5

MOVE_BAD = True


def is_single_hit_then_silent(y, sr=SR):
    pre_samples = int(PRE_MS / 1000.0 * sr)

    start = pre_samples + int(CHECK_START_MS / 1000.0 * sr)
    end = pre_samples + int(CHECK_END_MS / 1000.0 * sr)
    end = min(end, len(y))

    if start >= len(y) or start >= end:
        return True

    post = y[start:end]

    post_rms = np.sqrt(np.mean(post ** 2))
    active_ratio = np.mean(np.abs(post) > ACTIVE_AMP)

    if post_rms < MIN_POST_RMS and active_ratio < MIN_ACTIVE_RATIO:
        return True

    return False


def move_file(src_path, reason):
    rel = os.path.relpath(src_path, BASE_DIR)
    out_path = os.path.join(REJECT_DIR, reason, rel)
    os.makedirs(os.path.dirname(out_path), exist_ok=True)

    if MOVE_BAD:
        shutil.move(src_path, out_path)


def get_split_name(folder):
    rel = os.path.relpath(folder, BASE_DIR)
    split = rel.split(os.sep)[0]
    return split


def process_folder(folder):
    wav_files = sorted(glob.glob(os.path.join(folder, "*.wav")))
    if len(wav_files) == 0:
        return 0, 0, 0

    split = get_split_name(folder)
    limit = SPLIT_LIMITS.get(split, None)

    good_files = []
    removed_bad = 0
    removed_overflow = 0

    #Clean Bad Audio Clips
    for f in wav_files:
        y, _ = librosa.load(f, sr=SR, mono=True)

        bad = is_single_hit_then_silent(y)

        if bad:
            removed_bad += 1
            move_file(f, "bad_clips")
        else:
            good_files.append(f)

    #Remove Extra Clips Over Limit
    if limit is not None and len(good_files) > limit:
        kept_files = good_files[:limit]
        overflow_files = good_files[limit:]

        for f in overflow_files:
            removed_overflow += 1
            move_file(f, "overflow")

        kept = len(kept_files)
    else:
        kept = len(good_files)

    return kept, removed_bad, removed_overflow


total_kept = 0
total_removed_bad = 0
total_removed_overflow = 0

for root, dirs, files in os.walk(BASE_DIR):
    if any(f.endswith(".wav") for f in files):
        kept, removed_bad, removed_overflow = process_folder(root)

        total_kept += kept
        total_removed_bad += removed_bad
        total_removed_overflow += removed_overflow

        print(f"\n{root}")
        print(f"Kept: {kept}")
        print(f"Removed bad: {removed_bad}")
        print(f"Removed overflow: {removed_overflow}")

print("Total removed bad:", total_removed_bad)
print("Total removed overflow:", total_removed_overflow)
print("\n");

#Check File Size
for root, dirs, files in os.walk("extracted_audiosfinal"):
    wavs = [f for f in files if f.endswith(".wav")]
    if wavs:
        print(root, len(wavs))
```

# **3 FFT Feature Extraction**

```python
import pandas as pd

SR = 16000
N_FFT = 4096

DATA_ROOTS = {
    "train": "extracted_audiofinal/train",
    "validation": "extracted_audiofinal/validation",
    "test": "extracted_audiofinal/test",
}

BANDS = [
    (0, 250),
    (0, 500), #Added after Data Analysis
    (250, 500),
    (500, 1000),
    (1000, 2000),
    (2000, 3000),
    (3000, 4000),
    (4000, 6000),
    (6000, 8000),
]

def normalize_audio(y):
    peak = np.max(np.abs(y))
    if peak > 0:
        y = y / peak
    return y

def compute_fft_features(y, sr=SR, n_fft=N_FFT):
    y = normalize_audio(y)

    if len(y) < n_fft:
        y_pad = np.pad(y, (0, n_fft - len(y)))
    else:
        y_pad = y[:n_fft]

    window = np.hanning(len(y_pad))
    y_win = y_pad * window

    spectrum = np.abs(np.fft.rfft(y_win))
    freqs = np.fft.rfftfreq(len(y_pad), d=1 / sr)

    eps = 1e-10
    power = spectrum ** 2
    total_energy = np.sum(power) + eps

    peaks, _ = find_peaks(spectrum, distance=20)
    peak_mags = spectrum[peaks]

    if len(peaks) > 0:
        order = np.argsort(peak_mags)[::-1]
        sorted_peaks = peaks[order]
        top_freqs = freqs[sorted_peaks[:3]]
    else:
        top_freqs = np.array([0.0, 0.0, 0.0])

    #Peak Frequencies
    top_freqs = np.pad(top_freqs, (0, max(0, 3 - len(top_freqs))))[:3]

    #Spectral Centroid
    centroid = np.sum(freqs * spectrum) / (np.sum(spectrum) + eps)

    #Spectral Bandwidth
    bandwidth = np.sqrt(
        np.sum(((freqs - centroid) ** 2) * spectrum) / (np.sum(spectrum) + eps)
    )

    cumulative_energy = np.cumsum(power)

    #Spectral Rolloff 85%
    rolloff_85 = freqs[np.searchsorted(cumulative_energy, 0.85 * total_energy)]

    #Spectral Rolloff 95%
    rolloff_95 = freqs[np.searchsorted(cumulative_energy, 0.95 * total_energy)]

    #Spectral Flatness
    flatness = np.exp(np.mean(np.log(spectrum + eps))) / (np.mean(spectrum + eps))

    band_features = {}
    for low, high in BANDS:
        idx = np.where((freqs >= low) & (freqs < high))[0]
        band_energy = np.sum(power[idx]) / total_energy
        band_features[f"band_{low}_{high}"] = band_energy

    low_energy = (
        band_features["band_0_250"]
        + band_features["band_250_500"]
        + band_features["band_500_1000"]
    )

    mid_energy = (
        band_features["band_1000_2000"]
        + band_features["band_2000_3000"]
        + band_features["band_3000_4000"]
    )

    high_energy = (
        band_features["band_4000_6000"]
        + band_features["band_6000_8000"]
    )

    features = {
        "peak_freq_1": top_freqs[0],
        "peak_freq_2": top_freqs[1],
        "peak_freq_3": top_freqs[2],
        "spectral_centroid": centroid,
        "spectral_bandwidth": bandwidth,
        "spectral_rolloff_85": rolloff_85,
        "spectral_rolloff_95": rolloff_95,
        "spectral_flatness": flatness,
        "total_energy": total_energy,
        "low_mid_ratio": low_energy / (mid_energy + eps),
        "high_mid_ratio": high_energy / (mid_energy + eps),
        "low_high_ratio": low_energy / (high_energy + eps),
    }

    features.update(band_features)
    return features

def infer_label(path):
    path_lower = path.lower().replace("\\", "/")
    if "/bottle/" in path_lower:
        return "bottle"
    elif "/can/" in path_lower:
        return "can"
    return "unknown"

def build_feature_table(data_root, split_name):
    rows = []

    for root, dirs, files in os.walk(data_root):
        for fname in files:
            if not fname.lower().endswith(".wav"):
                continue

            fpath = os.path.join(root, fname)
            label = infer_label(fpath)

            y, _ = librosa.load(fpath, sr=SR, mono=True)
            feats = compute_fft_features(y, sr=SR, n_fft=N_FFT)

            row = {
                "filename": fname,
                "filepath": fpath,
                "split": split_name,
                "label": label,
            }
            row.update(feats)
            rows.append(row)

    return pd.DataFrame(rows)

#Build FFT Feature Tables
all_dfs = []

for split_name, data_root in DATA_ROOTS.items():
    df = build_feature_table(data_root, split_name)

    out_name = f"fft_feature_table_{split_name}_{N_FFT}final.csv"
    df.to_csv(out_name, index=False)

    print(f"\nSaved {out_name}: {df.shape}")
    print(df["label"].value_counts(dropna=False))

    all_dfs.append(df)
```

#**4 FFT Feature Analysis**

**4.1 Feature Effect Ranking**

```python
from sklearn.feature_selection import mutual_info_classif, f_classif
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("fft_feature_table_train_4096final.csv")

exclude_cols = ["label", "file", "segment_id"]
feature_cols = [c for c in df.columns if c not in exclude_cols and pd.api.types.is_numeric_dtype(df[c])]

# encode labels
le = LabelEncoder()
y = le.fit_transform(df["label"])

results = []

for col in feature_cols:
    x = df[col].values.astype(float)

    bottle_vals = df[df["label"] == "bottle"][col].dropna().values
    can_vals    = df[df["label"] == "can"][col].dropna().values

    mu_b = np.mean(bottle_vals)
    mu_c = np.mean(can_vals)
    std_b = np.std(bottle_vals, ddof=1)
    std_c = np.std(can_vals, ddof=1)

    # pooled_std
    pooled_std = np.sqrt((std_b**2 + std_c**2) / 2)
    effect_size = abs(mu_b - mu_c) / pooled_std if pooled_std > 0 else 0

    # single-feature ROC AUC
    try:
        auc = roc_auc_score(y, x)
        auc = max(auc, 1 - auc)  # irrelevant direction
    except:
        auc = np.nan

    results.append({
        "feature": col,
        "bottle_mean": mu_b,
        "can_mean": mu_c,
        "abs_diff": abs(mu_b - mu_c),
        "effect_size": effect_size,
        "single_feature_auc": auc
    })

result_df = pd.DataFrame(results)

# add mutual information and ANOVA F-score
X = df[feature_cols].fillna(0).values
mi = mutual_info_classif(X, y, random_state=42)
f_vals, p_vals = f_classif(X, y)

result_df["mutual_info"] = mi
result_df["f_score"] = f_vals
result_df["p_value"] = p_vals

result_df = result_df.sort_values(
    by=["effect_size", "single_feature_auc", "mutual_info"],
    ascending=False
)

print(result_df[[
    "feature",
    "effect_size",
    "single_feature_auc",
    "mutual_info",
    "f_score",
    "p_value",
    "abs_diff",
    "bottle_mean",
    "can_mean"
]].round(4))

result_df.to_csv("fft_feature_ranking.csv", index=False)
```

#**5 Model Training**

Models were trained using the following code implementation using the same random seed for reproducable results. Sessions are cleared before each model training.

```python
import random
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.utils import shuffle

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping

##################### Seed Setup ####################################

SEED = 67

random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

tf.keras.backend.clear_session()

########################## Load Files ##############################

train_df = pd.read_csv("fft_feature_table_train_4096final.csv")
val_df   = pd.read_csv("fft_feature_table_validation_4096final.csv")
test_df  = pd.read_csv("fft_feature_table_test_4096final.csv")

######################### Data Shuffling #############################

train_df = shuffle(train_df, random_state=SEED).reset_index(drop=True)
val_df   = shuffle(val_df, random_state=SEED).reset_index(drop=True)
test_df  = shuffle(test_df, random_state=SEED).reset_index(drop=True)

######################################################################

top_features = [
    "spectral_centroid",
    "spectral_flatness",
    "band_1000_2000",
    "band_0_500",
]

X_train = train_df[top_features].copy().values
X_val   = val_df[top_features].copy().values
X_test  = test_df[top_features].copy().values

y_train = train_df["label"].copy().values
y_val   = val_df["label"].copy().values
y_test  = test_df["label"].copy().values

#################### Label Encode ####################

le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_val_enc   = le.transform(y_val)
y_test_enc  = le.transform(y_test)

print("Label mapping:")
for cls, idx in zip(le.classes_, range(len(le.classes_))):
    print(cls, "->", idx)

print("\nCounts:")
print("Train:\n", train_df["label"].value_counts())
print("\nVal:\n", val_df["label"].value_counts())
print("\nTest:\n", test_df["label"].value_counts())


#################### Scale features ####################

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val   = scaler.transform(X_val)
X_test  = scaler.transform(X_test)

print("Train shape:", X_train.shape)
print("Val shape:", X_val.shape)
print("Test shape:", X_test.shape)

#################### Build Model ####################

model = Sequential([
    tf.keras.Input(shape=(X_train.shape[1],)),
    #Dense(4, activation="relu"),
    Dense(4, activation="relu"),
    Dense(4, activation="relu"),
    Dense(1, activation="sigmoid")
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=8,
    restore_best_weights=True
)

model.summary()

# Train
history = model.fit(
    X_train, y_train_enc,
    validation_data=(X_val, y_val_enc),
    epochs=1000,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)

# Test
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int).flatten()
```

**5.1 Plot Training VS Validation Loss and Accuracy / Reports**

```python
#################### Score Reports ####################

print("Test Accuracy:", accuracy_score(y_test_enc, y_pred))
print("Validation Accuracy:", max(history.history['val_accuracy']))
print("Train Accuracy:", max(history.history['accuracy']))

print("\nValidation Loss:", min(history.history['val_loss']))
print("Train Loss:", min(history.history['loss']))

print("\nClassification Report:")
print(classification_report(y_test_enc, y_pred, target_names=le.classes_))
print("Confusion Matrix:")
print(confusion_matrix(y_test_enc, y_pred))

#################### Train vs Validation Plot ####################

plt.figure(figsize=(8, 5))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.xlabel("Epoch")
plt.xlim([0, None])
plt.ylabel("Loss")
plt.ylim([0.0, 1])
plt.title("Training and Validation Loss (2048-4)")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Train Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.xlabel("Epoch")
plt.xlim([0, None])
plt.ylabel("Accuracy")
plt.ylim([0.1, 1.0])
plt.title("Training and Validation Accuracy (2048-4)")
plt.legend()
plt.grid(True)
plt.show()
```

#**TFLite Model Analysis**

```python
import time

MODEL_PATHS = {
    "INT8": "fft_bcr_int83.tflite",
    "DYNAMIC": "fft_bcr_dynamic3.tflite",
    "FLOAT32": "fft_bcr_float323.tflite",
}

x_sample = X_test[0:1].astype(np.float32)

def benchmark_tflite_model(model_name, model_path, x_sample, repeats=1000, warmup=50):
    interpreter = tf.lite.Interpreter(model_path=model_path)
    interpreter.allocate_tensors()

    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()

    input_index = input_details[0]["index"]
    output_index = output_details[0]["index"]

    input_dtype = input_details[0]["dtype"]
    output_dtype = output_details[0]["dtype"]

    in_scale, in_zero = input_details[0]["quantization"]
    out_scale, out_zero = output_details[0]["quantization"]

    if input_dtype == np.int8:
        x_in = np.round(x_sample / in_scale + in_zero).astype(np.int8)
        x_in = np.clip(x_in, -128, 127)
    else:
        x_in = x_sample.astype(input_dtype)

    # warmup
    for _ in range(warmup):
        interpreter.set_tensor(input_index, x_in)
        interpreter.invoke()
        _ = interpreter.get_tensor(output_index)

    # latency test
    times = []

    for _ in range(repeats):
        start = time.perf_counter()
        interpreter.set_tensor(input_index, x_in)
        interpreter.invoke()
        _ = interpreter.get_tensor(output_index)
        end = time.perf_counter()
        times.append((end - start) * 1000)  # ms

    # file size
    file_size_bytes = os.path.getsize(model_path)
    file_size_kb = file_size_bytes / 1024

    print(f"\n=== {model_name} ===")
    print("Model path:", model_path)
    print(f"Model size: {file_size_bytes} bytes ({file_size_kb:.2f} KB)")
    print("Input dtype:", input_dtype)
    print("Input quantization:", (in_scale, in_zero))
    print("Output dtype:", output_dtype)
    print("Output quantization:", (out_scale, out_zero))
    print(f"Average latency: {np.mean(times):.6f} ms")
    print(f"Min latency:     {np.min(times):.6f} ms")
    print(f"Max latency:     {np.max(times):.6f} ms")
    print(f"Std latency:     {np.std(times):.6f} ms")

    return {
        "model": model_name,
        "path": model_path,
        "size_bytes": file_size_bytes,
        "size_kb": file_size_kb,
        "input_dtype": str(input_dtype),
        "output_dtype": str(output_dtype),
        "avg_ms": np.mean(times),
        "min_ms": np.min(times),
        "max_ms": np.max(times),
        "std_ms": np.std(times),
    }

results = []

for name, path in MODEL_PATHS.items():
    results.append(benchmark_tflite_model(name, path, x_sample))

############### Summary ####################

print("\n\n===== SUMMARY =====")
print(f"{'Model':<10} {'Size (KB)':<12} {'Avg ms':<12} {'Min ms':<12} {'Max ms':<12} {'Std ms':<12}")
for r in results:
    print(f"{r['model']:<10} {r['size_kb']:<12.2f} {r['avg_ms']:<12.6f} {r['min_ms']:<12.6f} {r['max_ms']:<12.6f} {r['std_ms']:<12.6f}")
```

**7.2 Performance Analysis**

```python
############### Helpers ####################

def predict_tflite(model_path, X):
    interpreter = tf.lite.Interpreter(model_path=model_path)
    interpreter.allocate_tensors()

    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()

    input_index = input_details[0]["index"]
    output_index = output_details[0]["index"]

    input_dtype = input_details[0]["dtype"]
    output_dtype = output_details[0]["dtype"]

    in_scale, in_zero = input_details[0]["quantization"]
    out_scale, out_zero = output_details[0]["quantization"]

    preds = []

    for i in range(len(X)):
        x = X[i:i+1].astype(np.float32)

        # quantize input
        if input_dtype == np.int8:
            x_in = np.round(x / in_scale + in_zero).astype(np.int8)
        elif input_dtype == np.uint8:
            x_in = np.round(x / in_scale + in_zero).astype(np.uint8)
        else:
            x_in = x.astype(input_dtype)

        interpreter.set_tensor(input_index, x_in)
        interpreter.invoke()

        y = interpreter.get_tensor(output_index)

        # dequantize output
        if output_dtype == np.int8:
            y = (y.astype(np.float32) - out_zero) * out_scale
        elif output_dtype == np.uint8:
            y = (y.astype(np.float32) - out_zero) * out_scale
        else:
            y = y.astype(np.float32)

        preds.append(y[0][0])

    return np.array(preds)


############### Keras ####################

keras_probs = model.predict(X_test).flatten()
keras_pred = (keras_probs > 0.5).astype(int)

print("=== KERAS MODEL ===")
print("Accuracy:", accuracy_score(y_test_enc, keras_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test_enc, keras_pred))
print("Classification Report:")
print(classification_report(y_test_enc, keras_pred, target_names=le.classes_))


############### float32 ####################

float_probs = predict_tflite("fft_bcr_float323.tflite", X_test)
float_pred = (float_probs > 0.5).astype(int)

print("\n=== FLOAT32 TFLITE ===")
print("Accuracy:", accuracy_score(y_test_enc, float_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test_enc, float_pred))
print("Classification Report:")
print(classification_report(y_test_enc, float_pred, target_names=le.classes_))


############### Dynamic ####################

dynamic_probs = predict_tflite("fft_bcr_dynamic3.tflite", X_test)
dynamic_pred = (dynamic_probs > 0.5).astype(int)

print("\n=== DYNAMIC TFLITE ===")
print("Accuracy:", accuracy_score(y_test_enc, dynamic_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test_enc, dynamic_pred))
print("Classification Report:")
print(classification_report(y_test_enc, dynamic_pred, target_names=le.classes_))


############### int8 ####################

int8_probs = predict_tflite("fft_bcr_int83.tflite", X_test)
int8_pred = (int8_probs > 0.5).astype(int)

print("\n=== FULL INT8 TFLITE ===")
print("Accuracy:", accuracy_score(y_test_enc, int8_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test_enc, int8_pred))
print("Classification Report:")
print(classification_report(y_test_enc, int8_pred, target_names=le.classes_))

print("\n=== SUMMARY ===")
print(f"Keras Accuracy         : {accuracy_score(y_test_enc, keras_pred):.4f}")
print(f"Float32 TFLite Accuracy: {accuracy_score(y_test_enc, float_pred):.4f}")
print(f"Dynamic TFLite Accuracy: {accuracy_score(y_test_enc, dynamic_pred):.4f}")
print(f"INT8 TFLite Accuracy   : {accuracy_score(y_test_enc, int8_pred):.4f}")
```
